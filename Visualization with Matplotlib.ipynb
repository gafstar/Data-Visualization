# Two-Dimensional Histograms and Binnings

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mean=[0,0]
cov=[[1,1], [1,2]]
x, y=np.random.multivariate_normal(mean, cov, 10000).T
```

## Two Interfaces for the Price of One

Matplotlib has dual interfaces. One, which strictly imitates MATLAB-Style based interface and another more powerful and versatile OOP interface.
Following examples show the differences.

### MATLAB-Style Interface
The Matlab-style tools are contained in Pyplot(plt) interface. For instance:

```python
x=np.linspace(0, 3 * np.pi, 1000)
plt.figure() #create a plot figure

#create two plots
plt.subplot(2,1,1)  # (two rows, one col, 1st figure)
plt.plot(x, np.sin(x), color= 'pink', linestyle='-', linewidth= 2)
plt.xlim(-1, 11) # x-limit
plt.ylim(-1.5, 1.5)# y=limit
plt.title('A sine Curve')
plt.legend('sin(x)')

#create second panel
plt.subplot(2,1,2)
plt.plot(x, np.cos(x), color='blue', linestyle='-.', linewidth= 4.5)
plt.title('A Cosine Curve')
```

## OOP Style
In OO interface ot plotting, rather than calling the function (as opposed in MATLAB like plt.*), it is often more convenient to use *ax.set()* method to all these properties at once.

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
xlabel='x', ylabel='sin(x)',
title='A Simple Plot');
```

## Matlab-Style   Vs OOP Style

## MATLAB-Style               OOP Style
plt.xlabel()    -------->ax.set_xlabel()
plt.ylabel()    -------->ax.set_ylabel()
plt.xlim()      -------->ax.set_xlim()
plt.ylim()      -------->ax.set_ylim()
plt.title()     -------->ax.set_title()
